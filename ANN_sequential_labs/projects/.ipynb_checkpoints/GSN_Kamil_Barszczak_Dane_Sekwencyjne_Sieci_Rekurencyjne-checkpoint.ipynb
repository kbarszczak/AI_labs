```python
def plot_history(history):
    loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' not in s]
    val_loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' in s]
    acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' not in s]
    val_acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' in s]
    
    if len(loss_list) == 0:
        print('Loss is missing in history')
        return 
    
    ## As loss always exists
    epochs = range(1, len(history.history[loss_list[0]]) + 1)
    
    ## Loss
    plt.figure(1)
    for l in loss_list:
        plt.plot(epochs, history.history[l], 'b', label='Training loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    for l in val_loss_list:
        plt.plot(epochs, history.history[l], 'g', label='Validation loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    ## Accuracy
    plt.figure(2)
    for l in acc_list:
        plt.plot(epochs, history.history[l], 'b', label='Training accuracy (' + str(format(history.history[l][-1],'.5f'))+')')
    for l in val_acc_list:    
        plt.plot(epochs, history.history[l], 'g', label='Validation accuracy (' + str(format(history.history[l][-1],'.5f'))+')')

    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
def print_confusion_matrix(model, x_test, y_test, labels=["Negative", "Positive"]):
    y_pred = np.round(model.predict(x_test, verbose=0)).astype('int32')
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
def print_evaluation(model, x_test, y_test):
    loss, acc = model.evaluate(x_test, y_test, batch_size=batch_size, verbose=0)
    print(f"Test accuracy: {round(acc * 100, 2)}%")
    print(f"Test loss: {round(loss, 2)}")
```

```python
def load_data(path):
    # create list for labels & texts
    labels = []
    texts = []

    # loop though negative and positive reviews
    for label_type in ['neg', 'pos']:
        # create proper dir name
        dir_name = os.path.join(path, label_type)
        
        # loop though each file in the directory
        for fname in os.listdir(dir_name):
            # load only *.txt files
            if fname[-4:] == '.txt':
                f = open(os.path.join(dir_name, fname), encoding ="utf8")  # open file
                texts.append(f.read())  # read text
                f.close()  # close file
                if label_type == 'neg':  # append proper label
                    labels.append(0)
                else:
                    labels.append(1)
    
    # return the result
    return texts, labels
```

```python
def prepare_data(texts, labels, texts_test, labels_test, maxlen=None, max_words=10000):
    # create tokenizer
    tokenizer = Tokenizer(num_words=max_words)
    tokenizer.fit_on_texts(texts)
    
    # change texts to sequences of tokens
    sequences = tokenizer.texts_to_sequences(texts)
    sequences_test = tokenizer.texts_to_sequences(texts_test)

    # create sequences with equal lengths
    if maxlen is not None:
        data = pad_sequences(sequences, maxlen=maxlen)
        data_test = pad_sequences(sequences_test, maxlen=maxlen)
    else:
        data = pad_sequences(sequences)
        data_test = pad_sequences(sequences_test)

    # transform labels to a numpy array & return result for train and test dataset
    return data, np.asarray(labels), data_test, np.asarray(labels_test), tokenizer.word_index
```

```python
def split_data(data, labels, data_test, labels_test, coef=0.8):
    assert coef >= 0 and coef <= 1
    assert data.shape[0] == len(labels)
    assert data_test.shape[0] == len(labels_test)
    
    # create list of possible indies & shuffle it
    indices = np.arange(data.shape[0])
    np.random.shuffle(indices)
    
    indices_test = np.arange(data_test.shape[0])
    np.random.shuffle(indices_test)
    
    # shuffle the order of the data
    data = data[indices]
    labels = labels[indices]
    
    x_test = data_test[indices_test]
    y_test = labels_test[indices_test]

    # split data for training dataset
    training_samples = int(len(data) * coef)
    x_train = data[:training_samples]
    y_train = labels[:training_samples]
    
    # split data for validating dataset
    x_val = data[training_samples:]
    y_val = labels[training_samples:]
    
    # return the result
    return x_train, y_train, x_val, y_val, x_test, y_test
```

```python
def load_embedding_matrix(path, word_index, embedding_dim=200, max_words=10000):
    # create empty dict for the indices
    embeddings_index = {}
    
    # open file with the embedding matrix
    f = open(path, encoding ="utf8")
    
    # read line after line
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
        
    # close the file
    f.close()

    # create empty embedding matrix
    embedding_matrix = np.zeros((max_words, embedding_dim))
    for word, i in word_index.items():
        # read embedding vector
        embedding_vector = embeddings_index.get(word)
        
        # set up the embedding vector if the word is not out of the bound 
        if i < max_words:
            if embedding_vector is not None:
                # words not found in embedding index will be all-zeros.
                embedding_matrix[i] = embedding_vector
    
    # return the result
    return embedding_matrix
```

```python
# filenames
model_name = 'EmbeddingRNN_v4.h5'  # the model filename

# variables
maxlen = 500  # max sequence length - if 'None' then the maxlen parameter is equal to the longest sequence length
max_words = 10000  # max number of words in the dictionary
coef = 0.9  # training set split coefficient
embedding_dim = 25  # 2nd dimension size of the embedding matrix

# model params
epochs = 10
batch_size = 128
```

```python
# load data
print("Loading data...")
texts_train, labels_train = load_data(os.path.join(imdb_path, train_dir))
texts_test, labels_test = load_data(os.path.join(imdb_path, test_dir))

# prepare data
print("Preparing data...")
data_train, labels_train, data_test, labels_test, word_index = prepare_data(texts_train, labels_train, texts_test, labels_test, maxlen=maxlen, max_words=max_words)

# split data
print("Splitting data...")
x_train, y_train, x_val, y_val, x_test, y_test = split_data(data_train, labels_train, data_test, labels_test, coef=coef)
```

```python
# create sequential model
model = models.Sequential()

# embedding layer
model.add(layers.Embedding(max_words, embedding_dim, input_length=len(x_train[0])))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.2))

# RNN layer
model.add(layers.SimpleRNN(24, activation=layers.PReLU(), kernel_regularizer=regularizers.L1L2(l1=1e-2, l2=1e-2)))
model.add(layers.Dropout(0.1))

# Dense layer
model.add(layers.Dense(1, activation='sigmoid', kernel_regularizer=regularizers.L1L2(l1=1e-2, l2=1e-2)))

# compile model
model.compile(
    loss=losses.BinaryCrossentropy(),
    optimizer=optimizers.Nadam(learning_rate=0.001),
    metrics=['acc']
)

# print model summary
model.summary()
```

```python
# teach model
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=epochs, 
    batch_size=batch_size,
    validation_data=(x_val, y_val)
)
```

# Temperature data

_**General functions**_

```python
def read_temp_data(filename):
    file = open(filename)
    data = file.read()
    file.close()
    
    lines = data.split('\n')
    header = lines[0].split(',')
    lines = lines[1:]
    
    dim = len(header) - 1
    float_data = np.zeros((len(lines), dim))
    for i, line in enumerate(lines):
        values = [float(x) for x in line.split(',')[1:]]
        if len(values) == dim:
            float_data[i, :] = values

    mean = float_data.mean(axis=0)
    float_data -= mean
    std = float_data.std(axis=0)
    float_data /= std
    
    return float_data
```

```python
def generator(data, lookback, delay, min_index, max_index, shuffle=False, batch_size=128, step=6):
    if max_index is None:
        max_index = len(data) - delay - 1
    i = min_index + lookback
    while 1:
        if shuffle:
            rows = np.random.randint(min_index + lookback, max_index, size=batch_size)
        else:
            if i + batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i + batch_size, max_index))
            i += len(rows)

        samples = np.zeros((len(rows), lookback // step, data.shape[-1]))
        targets = np.zeros((len(rows),))
        for j, row in enumerate(rows):
            indices = range(rows[j] - lookback, rows[j], step)
            samples[j] = data[indices]
            targets[j] = data[rows[j] + delay][1]
        yield samples, targets
```

```python
def plot_last_loss_history(histories, n=None):
    colors = [('b', 'b.'), ('r', 'r.'), ('c', 'c.'), ('y', 'y.'), ('g', 'g.'), ('k', 'k.'), ('w', 'w.'), ('m', 'm.')]
    labels = [
        ('Training loss', 'Validation loss'), 
        ('Previous training loss', 'Previous validation loss'), 
        ('3rd training loss', '3rd validation loss'),
        ('4th training loss', '4th validation loss'),
        ('5th training loss', '5th validation loss'),
        ('6th training loss', '6th validation loss'),
        ('7th training loss', '7th validation loss'),
        ('8th training loss', '8th validation loss'),
    ]
    
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.figure(1)
    
    for i, history in enumerate(reversed(histories)):
        if (i >= 8) or (n is not None and i >= n):
            break
        
        color = colors[i][0]
        val_color = colors[i][1]
        
        label = labels[i][0]
        val_label = labels[i][1]
            
        loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' not in s]
        val_loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' in s]
        
        epochs = range(1, len(history.history[loss_list[0]]) + 1)
        
        for l in loss_list:
            plt.plot(epochs, history.history[l], color, label=label + ' (' + str(str(format(history.history[l][-1],'.5f'))+')'))
        for l in val_loss_list:
            plt.plot(epochs, history.history[l], val_color, label=val_label + ' (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    
    plt.legend()
    plt.show()
```

```python
def save_models(models, prefix):
    for index, model in enumerate(models):
        model.save(os.path.join(temp_model_path, prefix + '_' + str(index) + '.h5'))
```

```python
def load_models(path):
    models = []
    for file in os.scandir(path):
        path = file.path
        if re.search(".*\.h5$", path) is not None:
            model = keras.models.load_model(path)
            models.append(model)
```

```python
def test_loss(models, test_generator, steps):
    for index, model in enumerate(models):
        loss = model.evaluate(test_generator, steps=steps, batch_size=128, verbose=0)
        print(f"Test loss for model ({(index+1)}): {round(loss, 3)}")
```

### Podsumowanie

**NLP imbd:** najlepszy wynik przy problemie analizy sentymentu na zbiorze IMDB został uzyskany przez model złożony z warstwy Conv1D oraz warstwy Dense (model zawarty jest w poprzednim notebooku odesłanym przeze mnie jako poprzednie zadanie). Model ten uzyskał skuteczność na poziomie 89.27% oraz loss 0.39 na zbiorze testowym zawierającym 25000 próbek.

Wyniki dla modeli wykorzystujących warstwy rekurencyjne były bliskie wynikowi najlepszego modelu jednak w żadnym przypadku nie udało się przekroczyć progu 90%. Najlepszy wynik dla sieci rekurencyjnej to 87.87% skuteczności. Co ciekawe modele rekurencyjne zwykle osiągały większą skuteczność na zbiorze trenującym oraz walidującym niż modele nierekurencyjne jednak na zbiorze testowym wyniki były gorsze.

Przeprowadzone eksperymenty:
- Różna ilość gęstych warstw ukrytych (0, 1, 2): 2 i więcej warstw dawało gorsze wyniki niż 0 lub 1 warstwa ukryta. Model łatwiej było powstrzymać przed overfittingiem.
- Stosowanie LSTM oraz GRU: dla problemu nlp LSTM przynosił lepsze wyniki o około 1% skuteczności. GRU natomiast jest mniej skomplikowany i uczenie modelu było szybsze.
- Stosowanie LSTM oraz GRU jako warstwy ustawione na przemian po sobie: takie ułożenie okazało się być dość trudne do okiełznania pod kątem overfittingu.
- Różne funkcje aktywacji w warstwie ukrytej (PReLU, relu, tanh, softsign, elu): najlepsza okazała się być PReLU. Dobre efekty przynosi również funkcja tanh.
- Różne funkcje aktywacji w warstwach rekurencyjnych (activation: PReLU, relu, tanh, softsign; recurrent_activation: relu, sigmoid, softplus): jeśli chodzi o funkcje aktywacji to jedynymi sensownymi funkcjami okazały się być tanh oraz softsign (tanh dawał wyniki około 0.5% lepsze od softsign). Pozostałe funkcje powodowały problemy z eksplozją gradientu nawet w przypadku stosowania technik temu zapobiegających. Podobnie wygląda to w przypadku aktywacji rekurencyjnej, jedynymi sensownymi funkcjami były sigmoid oraz softplus (sigmoid dawał lepsze wyniki).
- Stosowanie warstw konwolucyjnych przed rekurencyjnymi: w większości przypadków przynosiło to dobry efekty, w szczególności z połączeniem z warstwą AveragePooling. Jednak tutaj nie można przesadzić z rozmiarem warstwy konwolucyjnej, ponieważ powoduje to bardzo duży overfitting
- Stosowanie wielu warstw rekurencyjnych (2, 3, 4): nie przynosiło to nic innego oprócz overfittingu. 1 warstwa rekurencyjna była w zupełności wystarczająca.
- Stosowanie dropoutu rekurencyjnego: znacznie zmniejszało to overfitting i w niektórych przypadkach polepszało wyniki skuteczności. Jednak ze względu na brak wsparcia dla GPU test wykonałem na komputerze z mocniejszym CPU przez co wyniki nie są pokazane w tym notebooku.
- Stosowanie regularyzacji z różnymi wartościami parametrów l1 i l2: przy odpowiednim dobraniu parametrów model osiąga lepsze wyniki skuteczności. Łatwo jest jednak przesadzić bądź dobrać zbyt małe wartości parametrów co pogarsza wyniki. Tutaj trzeba przeprowadzić szereg testów i wybrać najlepiej prosperującą wartość.

Wnioski:
- Dobre efekty daje następujące połączenie warstw: Wejście -> Embedding -> Conv1D -> AveragePooling -> GRU/LSTM -> Dense -> Dense (wyjście)
- Modele z niewielką ilością filtrów i jednostek w LSTM/GRU osiągają lepsze wyniki od modeli bardziej złożonych. W dodatku znaczenie szybciej przebiega uczenie takich modeli oraz znacznie łatwiej jest się uporać z overfittigiem.
- Wielkość dropout'u oraz parametrów regularyzacji może znacznie zmienić przebieg uczenia. Ich odpowiednie dobranie zwiększa skuteczność nawet o 2% i jednocześnie eliminuje problem overfitting'u.
- Inne niż domyślne (z wyjątkiem softsign oraz softplus) funkcje aktywacji w warstwach rekurencyjnych powodują jedynie problemy z eksplozją gradientu. Nie wpływają one pozytywnie na wyniki uczenia.
