_**General functions**_

```python
def indices2words(vec, word_index):
    index_word = {v: k for k, v in word_index.items()}
    result = ""
    for i, v in enumerate(vec):
        if v in index_word:
            result += index_word[v]
            if i != len(word_index)-1:
                result += " "
    return result
```

```python
def plot_history(history):
    loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' not in s]
    val_loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' in s]
    acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' not in s]
    val_acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' in s]
    
    if len(loss_list) == 0:
        print('Loss is missing in history')
        return 
    
    ## As loss always exists
    epochs = range(1, len(history.history[loss_list[0]]) + 1)
    
    ## Loss
    plt.figure(1)
    for l in loss_list:
        plt.plot(epochs, history.history[l], 'b', label='Training loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    for l in val_loss_list:
        plt.plot(epochs, history.history[l], 'g', label='Validation loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    ## Accuracy
    plt.figure(2)
    for l in acc_list:
        plt.plot(epochs, history.history[l], 'b', label='Training accuracy (' + str(format(history.history[l][-1],'.5f'))+')')
    for l in val_acc_list:    
        plt.plot(epochs, history.history[l], 'g', label='Validation accuracy (' + str(format(history.history[l][-1],'.5f'))+')')

    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
def print_confusion_matrix(model, x_test, y_test, labels=["Negative", "Positive"]):
    y_pred = np.round(model.predict(x_test, verbose=0)).astype('int32')
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
def print_evaluation(model, x_test, y_test):
    loss, acc = model.evaluate(x_test, y_test, batch_size=batch_size, verbose=0)
    print(f"Test accuracy: {round(acc * 100, 2)}%")
    print(f"Test loss: {round(loss, 2)}")
```

```python
def load_data(path):
    # create list for labels & texts
    labels = []
    texts = []

    # loop though negative and positive reviews
    for label_type in ['neg', 'pos']:
        # create proper dir name
        dir_name = os.path.join(path, label_type)
        
        # loop though each file in the directory
        for fname in os.listdir(dir_name):
            # load only *.txt files
            if fname[-4:] == '.txt':
                f = open(os.path.join(dir_name, fname), encoding ="utf8")  # open file
                texts.append(f.read())  # read text
                f.close()  # close file
                if label_type == 'neg':  # append proper label
                    labels.append(0)
                else:
                    labels.append(1)
    
    # return the result
    return texts, labels
```

```python
def prepare_data(texts, labels, texts_test, labels_test, maxlen=None, max_words=10000):
    # create tokenizer
    tokenizer = Tokenizer(num_words=max_words)
    tokenizer.fit_on_texts(texts)
    
    # change texts to sequences of tokens
    sequences = tokenizer.texts_to_sequences(texts)
    sequences_test = tokenizer.texts_to_sequences(texts_test)

    # create sequences with equal lengths
    if maxlen is not None:
        data = pad_sequences(sequences, maxlen=maxlen)
        data_test = pad_sequences(sequences_test, maxlen=maxlen)
    else:
        data = pad_sequences(sequences)
        data_test = pad_sequences(sequences_test)

    # transform labels to a numpy array & return result for train and test dataset
    return data, np.asarray(labels), data_test, np.asarray(labels_test), tokenizer.word_index
```

```python
def split_data(data, labels, data_test, labels_test, coef=0.8):
    assert coef >= 0 and coef <= 1
    assert data.shape[0] == len(labels)
    assert data_test.shape[0] == len(labels_test)
    
    # create list of possible indies & shuffle it
    indices = np.arange(data.shape[0])
    np.random.shuffle(indices)
    
    indices_test = np.arange(data_test.shape[0])
    np.random.shuffle(indices_test)
    
    # shuffle the order of the data
    data = data[indices]
    labels = labels[indices]
    
    x_test = data_test[indices_test]
    y_test = labels_test[indices_test]

    # split data for training dataset
    training_samples = int(len(data) * coef)
    x_train = data[:training_samples]
    y_train = labels[:training_samples]
    
    # split data for validating dataset
    x_val = data[training_samples:]
    y_val = labels[training_samples:]
    
    # return the result
    return x_train, y_train, x_val, y_val, x_test, y_test
```

```python
def load_embedding_matrix(path, word_index, embedding_dim=200, max_words=10000):
    # create empty dict for the indices
    embeddings_index = {}
    
    # open file with the embedding matrix
    f = open(path, encoding ="utf8")
    
    # read line after line
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
        
    # close the file
    f.close()

    # create empty embedding matrix
    embedding_matrix = np.zeros((max_words, embedding_dim))
    for word, i in word_index.items():
        # read embedding vector
        embedding_vector = embeddings_index.get(word)
        
        # set up the embedding vector if the word is not out of the bound 
        if i < max_words:
            if embedding_vector is not None:
                # words not found in embedding index will be all-zeros.
                embedding_matrix[i] = embedding_vector
    
    # return the result
    return embedding_matrix
```

### Model v1

_**Model parameters**_

```python
# filenames
model_name = 'EmbeddingRNN_v1_4.h5'  # the model filename

# variables
maxlen = 500  # max sequence length - if 'None' then the maxlen parameter is equal to the longest sequence length
max_words = 10000  # max number of words in the dictionary
coef = 0.9  # training set split coefficient
embedding_dim = 25  # 2nd dimension size of the embedding matrix

# model params
epochs = 20
batch_size = 128
```

_**Loading & processing data for the model**_

```python
# load data
print("Loading data...")
texts_train, labels_train = load_data(os.path.join(imdb_path, train_dir))
texts_test, labels_test = load_data(os.path.join(imdb_path, test_dir))

# prepare data
print("Preparing data...")
data_train, labels_train, data_test, labels_test, word_index = prepare_data(texts_train, labels_train, texts_test, labels_test, maxlen=maxlen, max_words=max_words)

# split data
print("Splitting data...")
x_train, y_train, x_val, y_val, x_test, y_test = split_data(data_train, labels_train, data_test, labels_test, coef=coef)
```

```python
print("Data info:")
print("Train size -", len(x_train))
print("Test size -", len(x_test))
print("Valid size -", len(x_val))
```

_**Creating & teaching model**_

```python
# create sequential model
model = models.Sequential([
    # embedding layer
    layers.Embedding(max_words, embedding_dim, input_length=len(x_train[0])),
    layers.BatchNormalization(),
    
    # flatten 
    layers.Flatten(),
    layers.Dropout(0.5),
    
    # 1st dense layer
    layers.Dense(24, activation=layers.PReLU(), kernel_regularizer=regularizers.L1L2(l1=1e-2, l2=1e-2)),
    layers.Dropout(0.25),
    
    # last layer
    layers.Dense(1, activation='sigmoid', kernel_regularizer=regularizers.L1L2(l1=1e-2, l2=1e-2))
])

# compile model
model.compile(
    loss=losses.BinaryCrossentropy(),
    optimizer=optimizers.Nadam(learning_rate=0.001),
    metrics=['acc']
)

# print model summary
model.summary()
```

```python
# teach model
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=epochs, 
    batch_size=batch_size,
    validation_data=(x_val, y_val),
#     callbacks=[
#         callbacks.ModelCheckpoint(os.path.join(model_path, model_name), save_freq="epoch", verbose=0)
#     ]
)
```

```python
model.save_weights(os.path.join(model_path, model_name))
```

```python
plot_history(history)
```

```python
print_evaluation(model, x_test, y_test)
```

```python
print_confusion_matrix(model, x_test, y_test)
```

### Stworzenie modelu 2

```python
# filenames
model_name = 'EmbeddingRNN_v2_3.h5'  # the model filename
glove_file = 'glove.6B.300d.txt'  # the embedding matrix file name

# variables
maxlen = 1000  # max sequence length - if 'None' then the maxlen parameter is equal to the longest sequence length
max_words = 15000  # max number of words in the dictionary
coef = 0.9  # training set split coefficient
embedding_dim = 300  # 2nd dimension size of the embedding matrix

# model params
epochs = 25
batch_size = 128
```

```python
# load data
print("Loading data...")
texts_train, labels_train = load_data(os.path.join(imdb_path, train_dir))
texts_test, labels_test = load_data(os.path.join(imdb_path, test_dir))

# prepare data
print("Preparing data...")
data_train, labels_train, data_test, labels_test, word_index = prepare_data(texts_train, labels_train, texts_test, labels_test, maxlen=maxlen, max_words=max_words)

# split data
print("Splitting data...")
x_train, y_train, x_val, y_val, x_test, y_test = split_data(data_train, labels_train, data_test, labels_test, coef=coef)

# read embedding matrix
print("Reading embedding matrix...")
embedding_matrix = load_embedding_matrix(os.path.join(glove_path, glove_file), word_index, embedding_dim=embedding_dim, max_words=max_words)
```

```python
print("Data info:")
print("Train size -", len(x_train))
print("Test size -", len(x_test))
print("Valid size -", len(x_val))
```

```python
# create sequential model
model = models.Sequential([
    # embedding layer
    layers.Embedding(max_words, embedding_dim, input_length=len(x_train[0])),
    
    # 1st convolutional layer
    layers.Conv1D(64, 3, activation=layers.PReLU(), padding='same', kernel_regularizer=regularizers.L1L2(l1=1e-4, l2=5e-4)),
    layers.AveragePooling1D(3),
    layers.Dropout(0.5),
    
    # 2nd convolutional layer
    layers.Conv1D(64, 3, activation=layers.PReLU(), padding='same', kernel_regularizer=regularizers.L1L2(l1=1e-4, l2=5e-4)),
    layers.AveragePooling1D(3),
    layers.Dropout(0.5),
    
    # 3rd convolutional layer
    layers.Conv1D(64, 3, activation=layers.PReLU(), padding='same', kernel_regularizer=regularizers.L1L2(l1=1e-4, l2=5e-4)),
    layers.AveragePooling1D(3),
    layers.Dropout(0.5),
    
    # flatten
    layers.Flatten(),
    
    # 1st dense layer
    layers.Dense(48, activation=layers.PReLU(), kernel_regularizer=regularizers.L1L2(l1=1e-4, l2=5e-4)),
    layers.Dropout(0.5),

    # last layer
    layers.Dense(1, activation='sigmoid', kernel_regularizer=regularizers.L1L2(l1=1e-4, l2=5e-4))
])

# set up embedding layer weights
model.layers[0].set_weights([embedding_matrix])
model.layers[0].trainable = False

# compile model
model.compile(
    loss=losses.BinaryCrossentropy(),
    optimizer=optimizers.Nadam(learning_rate=0.0005),
    metrics=['acc']
)

# print model summary
model.summary()
```

```python
### teach model
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=epochs, 
    batch_size=batch_size,
    validation_data=(x_val, y_val),
)
```

```python
plot_history(history)
```

```python
print_evaluation(model, x_test, y_test)
```

```python
print_confusion_matrix(model, x_test, y_test)
```

### Stworzenie modelu 3

```python
# filenames
model_name = 'EmbeddingRNN_v3_1.h5'  # the model filename
glove_file = 'glove.6B.300d.txt'  # the embedding matrix file name

# variables
maxlen = 500  # max sequence length - if 'None' then the maxlen parameter is equal to the longest sequence length
max_words = 15000  # max number of words in the dictionary
coef = 0.9  # training set split coefficient
embedding_dim = 300  # 2nd dimension size of the embedding matrix

# model params
epochs = 20
batch_size = 128
```

```python
# load data
print("Loading data...")
texts_train, labels_train = load_data(os.path.join(imdb_path, train_dir))
texts_test, labels_test = load_data(os.path.join(imdb_path, test_dir))

# prepare data
print("Preparing data...")
data_train, labels_train, data_test, labels_test, word_index = prepare_data(texts_train, labels_train, texts_test, labels_test, maxlen=maxlen, max_words=max_words)

# split data
print("Splitting data...")
x_train, y_train, x_val, y_val, x_test, y_test = split_data(data_train, labels_train, data_test, labels_test, coef=coef)

# read embedding matrix
print("Reading embedding matrix...")
embedding_matrix = load_embedding_matrix(os.path.join(glove_path, glove_file), word_index, embedding_dim=embedding_dim, max_words=max_words)
```

```python
print("Data info:")
print("Train size -", len(x_train))
print("Test size -", len(x_test))
print("Valid size -", len(x_val))
```

```python
# create sequential model
model = models.Sequential([
    # embedding layer
    layers.Embedding(max_words, embedding_dim, input_length=len(x_train[0])),
    layers.Dropout(0.5),
    
    # 1st convolutional layer
    layers.Conv1D(64, 3, activation='tanh', padding='same', kernel_regularizer=regularizers.L1L2(l1=3e-4, l2=7e-4)),
    layers.AveragePooling1D(3),
    layers.Dropout(0.25),
    
    # 2nd convolutional layer
    layers.Conv1D(64, 3, activation='tanh', padding='same', kernel_regularizer=regularizers.L1L2(l1=3e-4, l2=7e-4)),
    layers.AveragePooling1D(3),
    layers.Dropout(0.25),
    
    # flatten
    layers.Flatten(),
    
    # 1st dense layer
    layers.Dense(48, activation='tanh', kernel_regularizer=regularizers.L1L2(l1=3e-4, l2=7e-4)),
    layers.Dropout(0.25),

    # last layer
    layers.Dense(1, activation='sigmoid')
])

# set up embedding layer weights
model.layers[0].set_weights([embedding_matrix])

# compile model
model.compile(
    loss=losses.BinaryCrossentropy(),
    optimizer=optimizers.Nadam(learning_rate=0.0005),
    metrics=['acc']
)

# print model summary
model.summary()
```

```python
### teach model
model.layers[0].trainable=False
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=int(epochs/3), 
    batch_size=batch_size,
    validation_data=(x_val, y_val),
)

model.layers[0].trainable=True
history2 = model.fit(
    x=x_train,
    y=y_train,
    epochs=int(epochs/3*2), 
    batch_size=batch_size,
    validation_data=(x_val, y_val),
)

for key, item in history.history.items():
    item.extend(history2.history[key])
```

#### Model 1
Skuteczność na zbiorze testowym: 86.45%. Model ten wykorzystuje 1 warstwę Embedding oraz Dense. Przeprowadzone testy:
- Długość frazy wejściowej (100, 200, 400, 500, 1000, 2000) - testy pokazały że im fraza była dłuższa tym model osiągał lepsze wyniki. Jednak po przekroczeniu pewnego progu (ok. 1000) model stał się skomplikowany i przestał się skutecznie uczyć. Powstał duży problem z overfittingiem.
- Zastosowanie embeddingu o różnym rozmiarze (25, 50, 100, 200, 300) - większy wymiar embeddingu znacznie komplikował model. Wymiar 25 okazał się być najlepszy i dał najwyższą skuteczność
- Wykorzystanie różnych funkcji aktywacji (relu, selu, elu, leakyrelu, prelu, tanh, sigmoid) - wyniki testów przedstawione są na poniższym obrazku:

- Warstwa dense z różną ilością neuronów (4, 8, 12, 16, 24, 32, 36, 48, 64, 128) - zwiększenie liczby neuronów znacznie komplikowało model oraz zwiększało problemy overfittingu. Natomiast zbyt mała ilość neuronów powodowała, że model nie był w stanie się wszystkiego nauczyć i osiągane wyniki były kiepskie.
- Różna ilość warstw dense (2, 3, 4) - przeprowadzone testy potwierdziły, że zwiększenie ilość warstw ukrytych nie miało pozytywnego wpływu na osiągane wyniki. Najlepsza skuteczność była osiągana przez model z 1 warstwą ukrytą.
- Zastosowanie regularyzacji oraz dropout - odpowiednie dobranie parametrów regularyzacji okazało się być kluczem. Zbyt duża wartość tych parametrów uniemożliwiała uczenie, natomiast zbyt mała nie chroniła przed overfittingiem. Po wielu testach udało się jednak dobrać wartość, która poprawiła wyniki uczenia.

#### Model 2
Skuteczność na zbiorze testowym: 89.27%. Model wykorzystuje sekwencje o długości 1000 oraz wymiar osadzania 300. Dla warstwy embedding wykorzystane zostały wagi glove.6B.300d. Model składa się z 3 jednowymiarowych warstw konwolucyjnych, po których występuje AveragePooling. Po warstwie Flatten występuje tylko 1 ukryta warstwa Dense. Oprócz testów podobnych do tych z modelu 1, zostały również przeprowadzone następujące testy:
- Zastosowanie AveragePooling oraz MaxPooling - AveragePooling generuje wyższą skuteczność
- Modyfikowanie liczby filtrów w warstwach konwolucyjnych (3, 4, 5, 8, 12, 16, 32, 64, 128, 256) - zwiększenie liczby filtrów polepszało wyniki jednak pojawiał się większy problem z overfittingiem. Dla zbyt dużej liczby pojawiał się również problem wybuchających gradientów
- Modyfikowanie rozmiaru filtra konwolucyjnego (2, 3, 4, 5, 7, 8, 10, 12, 20, 50) - zmiana filtra na większy zmniejszała skuteczność uczenia modelu. Najlepsze wartości okazały się być w przedziale od 3 do 5. Poniżej rysunek przedstawiający historię uczenia modelu bez oraz z regularyzacja:
