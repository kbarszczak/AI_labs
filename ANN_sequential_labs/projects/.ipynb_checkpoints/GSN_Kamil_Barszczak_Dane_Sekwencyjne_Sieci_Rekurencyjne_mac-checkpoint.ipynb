```python
def plot_history(history):
    loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' not in s]
    val_loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' in s]
    acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' not in s]
    val_acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' in s]
    
    if len(loss_list) == 0:
        print('Loss is missing in history')
        return 
    
    ## As loss always exists
    epochs = range(1, len(history.history[loss_list[0]]) + 1)
    
    ## Loss
    plt.figure(1)
    for l in loss_list:
        plt.plot(epochs, history.history[l], 'b', label='Training loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    for l in val_loss_list:
        plt.plot(epochs, history.history[l], 'g', label='Validation loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    ## Accuracy
    plt.figure(2)
    for l in acc_list:
        plt.plot(epochs, history.history[l], 'b', label='Training accuracy (' + str(format(history.history[l][-1],'.5f'))+')')
    for l in val_acc_list:    
        plt.plot(epochs, history.history[l], 'g', label='Validation accuracy (' + str(format(history.history[l][-1],'.5f'))+')')

    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
def print_confusion_matrix(model, x_test, y_test, labels=["Negative", "Positive"]):
    y_pred = np.round(model.predict(x_test, verbose=0)).astype('int32')
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
```

```python
def print_evaluation(model, x_test, y_test):
    loss, acc = model.evaluate(x_test, y_test, batch_size=batch_size, verbose=0)
    print(f"Test accuracy: {round(acc * 100, 2)}%")
    print(f"Test loss: {round(loss, 2)}")
```

```python
def load_data(path):
    # create list for labels & texts
    labels = []
    texts = []

    # loop though negative and positive reviews
    for label_type in ['neg', 'pos']:
        # create proper dir name
        dir_name = os.path.join(path, label_type)
        
        # loop though each file in the directory
        for fname in os.listdir(dir_name):
            # load only *.txt files
            if fname[-4:] == '.txt':
                f = open(os.path.join(dir_name, fname), encoding ="utf8")  # open file
                texts.append(f.read())  # read text
                f.close()  # close file
                if label_type == 'neg':  # append proper label
                    labels.append(0)
                else:
                    labels.append(1)
    
    # return the result
    return texts, labels
```

```python
def prepare_data(texts, labels, texts_test, labels_test, maxlen=None, max_words=10000):
    # create tokenizer
    tokenizer = Tokenizer(num_words=max_words)
    tokenizer.fit_on_texts(texts)
    
    # change texts to sequences of tokens
    sequences = tokenizer.texts_to_sequences(texts)
    sequences_test = tokenizer.texts_to_sequences(texts_test)

    # create sequences with equal lengths
    if maxlen is not None:
        data = pad_sequences(sequences, maxlen=maxlen)
        data_test = pad_sequences(sequences_test, maxlen=maxlen)
    else:
        data = pad_sequences(sequences)
        data_test = pad_sequences(sequences_test)

    # transform labels to a numpy array & return result for train and test dataset
    return data, np.asarray(labels), data_test, np.asarray(labels_test), tokenizer.word_index
```

```python
def split_data(data, labels, data_test, labels_test, coef=0.8):
    assert coef >= 0 and coef <= 1
    assert data.shape[0] == len(labels)
    assert data_test.shape[0] == len(labels_test)
    
    # create list of possible indies & shuffle it
    indices = np.arange(data.shape[0])
    np.random.shuffle(indices)
    
    indices_test = np.arange(data_test.shape[0])
    np.random.shuffle(indices_test)
    
    # shuffle the order of the data
    data = data[indices]
    labels = labels[indices]
    
    x_test = data_test[indices_test]
    y_test = labels_test[indices_test]

    # split data for training dataset
    training_samples = int(len(data) * coef)
    x_train = data[:training_samples]
    y_train = labels[:training_samples]
    
    # split data for validating dataset
    x_val = data[training_samples:]
    y_val = labels[training_samples:]
    
    # return the result
    return x_train, y_train, x_val, y_val, x_test, y_test
```

```python
def load_embedding_matrix(path, word_index, embedding_dim=200, max_words=10000):
    # create empty dict for the indices
    embeddings_index = {}
    
    # open file with the embedding matrix
    f = open(path, encoding ="utf8")
    
    # read line after line
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
        
    # close the file
    f.close()

    # create empty embedding matrix
    embedding_matrix = np.zeros((max_words, embedding_dim))
    for word, i in word_index.items():
        # read embedding vector
        embedding_vector = embeddings_index.get(word)
        
        # set up the embedding vector if the word is not out of the bound 
        if i < max_words:
            if embedding_vector is not None:
                # words not found in embedding index will be all-zeros.
                embedding_matrix[i] = embedding_vector
    
    # return the result
    return embedding_matrix
```

# IMDB with recursion

### SimpleRNN

```python
# filenames
model_name = 'EmbeddingRNN_v4.h5'  # the model filename

# variables
maxlen = 500  # max sequence length - if 'None' then the maxlen parameter is equal to the longest sequence length
max_words = 10000  # max number of words in the dictionary
coef = 0.9  # training set split coefficient
embedding_dim = 25  # 2nd dimension size of the embedding matrix

# model params
epochs = 10
batch_size = 128
```

```python
# load data
print("Loading data...")
texts_train, labels_train = load_data(os.path.join(imdb_path, train_dir))
texts_test, labels_test = load_data(os.path.join(imdb_path, test_dir))

# prepare data
print("Preparing data...")
data_train, labels_train, data_test, labels_test, word_index = prepare_data(texts_train, labels_train, texts_test, labels_test, maxlen=maxlen, max_words=max_words)

# split data
print("Splitting data...")
x_train, y_train, x_val, y_val, x_test, y_test = split_data(data_train, labels_train, data_test, labels_test, coef=coef)
```

```python
print("Data info:")
print("Train size -", len(x_train))
print("Test size -", len(x_test))
print("Valid size -", len(x_val))
```

```python
# create sequential model
model = models.Sequential()

# embedding layer
model.add(layers.Embedding(max_words, embedding_dim, input_length=len(x_train[0])))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.2))

# RNN layer
model.add(layers.SimpleRNN(24, activation=layers.PReLU(), kernel_regularizer=regularizers.L1L2(l1=1e-2, l2=1e-2)))
model.add(layers.Dropout(0.1))

# Dense layer
model.add(layers.Dense(1, activation='sigmoid', kernel_regularizer=regularizers.L1L2(l1=1e-2, l2=1e-2)))

# compile model
model.compile(
    loss=losses.BinaryCrossentropy(),
    optimizer=optimizers.Nadam(learning_rate=0.001),
    metrics=['acc']
)

# print model summary
model.summary()
```

```python
# teach model
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=epochs, 
    batch_size=batch_size,
    validation_data=(x_val, y_val)
)
```

```python
model.save_weights(os.path.join(imdb_model_path, model_name))
```

```python
plot_history(history)
```

```python
print_evaluation(model, x_test, y_test)
```

```python
print_confusion_matrix(model, x_test, y_test)
```

### GRU

```python
# filenames
model_name = 'EmbeddingRNN_v6.h5'  # the model filename

# variables
maxlen = 500  # max sequence length - if 'None' then the maxlen parameter is equal to the longest sequence length
max_words = 15000  # max number of words in the dictionary
coef = 0.9  # training set split coefficient
embedding_dim = 100  # 2nd dimension size of the embedding matrix

# model params
epochs = 10
batch_size = 128
```

```python
# load data
print("Loading data...")
texts_train, labels_train = load_data(os.path.join(imdb_path, train_dir))
texts_test, labels_test = load_data(os.path.join(imdb_path, test_dir))

# prepare data
print("Preparing data...")
data_train, labels_train, data_test, labels_test, word_index = prepare_data(texts_train, labels_train, texts_test, labels_test, maxlen=maxlen, max_words=max_words)

# split data
print("Splitting data...")
x_train, y_train, x_val, y_val, x_test, y_test = split_data(data_train, labels_train, data_test, labels_test, coef=coef)
```

```python
print("Data info:")
print("Train size -", len(x_train))
print("Test size -", len(x_test))
print("Valid size -", len(x_val))
```

```python
l1 = 5e-3
l2 = 5e-3
dropout = 0.2

# create sequential model
model = models.Sequential()

# embedding layer
model.add(layers.Embedding(max_words, embedding_dim, input_length=len(x_train[0])))
model.add(layers.Dropout(dropout))
model.add(layers.BatchNormalization())

# 1st Conv1D & pooling layer
model.add(layers.Conv1D(8, 3, activation=layers.PReLU(), kernel_regularizer=regularizers.L1L2(l1=l1, l2=l2)))
model.add(layers.AveragePooling1D(3))
model.add(layers.Dropout(dropout))

# 1st LSTM layer
model.add(layers.GRU(32, activation='tanh', recurrent_activation='sigmoid', dropout=dropout, kernel_regularizer=regularizers.L1L2(l1=l1, l2=l2)))
model.add(layers.Dropout(dropout))

# 1st Dense layer
model.add(layers.Dense(16, activation=layers.PReLU(), kernel_regularizer=regularizers.L1L2(l1=l1, l2=l2)))
model.add(layers.Dropout(dropout))

# final layer
model.add(layers.Dense(1, activation='sigmoid', kernel_regularizer=regularizers.L1L2(l1=l1, l2=l2)))
          
# compile model
model.compile(
    loss=losses.BinaryCrossentropy(),
    optimizer=optimizers.Nadam(learning_rate=0.001),
    metrics=['acc']
)

# print model summary
model.summary()
```

```python
### teach model
history = model.fit(
    x=x_train,
    y=y_train,
    epochs=epochs, 
    batch_size=batch_size,
    validation_data=(x_val, y_val),
)
```

```python
plot_history(history)
```

```python
print_evaluation(model, x_test, y_test)
```

```python
print_confusion_matrix(model, x_test, y_test)
```
